## Purpose of Analysis
___

The purpose of this work is to evaluate D.Trump's tweets and their effect on the stock market and on his approval rating. We aim to identify if there is a relationship between these variables. 

**Why It Matters?**
___

It is important to know how much impact Trumps tweets are making and how much attention you have to pay or be worried about it or not worry about it at all.

**Used Terminology**
_____

* **P-value** - In statistical hypothesis testing, the p-value or probability value is, for a given statistical model, the probability that, when the null hypothesis is true, the statistical summary would be equal to, or more extreme than, the actual observed results

* **Mean** - The statistical mean refers to the mean or average that is used to derive the central tendency of the data in question

* **t-statistics** - The t-statistic is the ratio of the departure of the estimated value of a parameter from its hypothesized value to its standard error

* **ETF** - Exchange-Traded Fund

To represent the stock market we used Vanguard's SP500 ETF (ticker:VOO)

**Data**
____

* VOO ETF from Alpha Vantage (www.alphavantage.co) 
* Trump's Tweet Archive - (2017-18)
* Trump's Approval Rating - (2017-18)

## Visualisations and Results##

## Conclusions: ##

1. There is a negative correlation between Trump's Tweets and the stock market prices. The Trump tweets, more it affects stock market prices negatively.

2. There is no relationship between:
